1. 쇼핑몰의 배송료를 계산하는 코드작성
- 사용하는 변수 : price, shipping_cost

```python
print("상품의 가격을 입력하시오.")
price = int(input("구입금액 입력: "))

if price > 20000:
    shipping_cost = 0  # 구입금액이 2만원 이상일 경우 배송비를 0으로 설정합니다.
    print(f'구입금액이 2만원 이상입니다.')
else:
    shipping_cost = 3000  # 구입금액이 2만원 이하일 경우 배송비를 3000으로 설정합니다.
    print(f'구입금액이 2만원 이하입니다.')

print(f'회원님의 배송비는 {shipping_cost}원입니다.')
```

위 코드는 사용자로부터 구입금액을 입력받아, 해당 가격에 따라 배송비를 계산하여 출력하는 코드입니다.

price 변수는 사용자로부터 입력받은 구입금액을 저장하는 변수입니다.
if 문을 사용하여 구입금액이 2만원 이상인지 아닌지를 확인합니다.
price > 20000은 구입금액이 2만원보다 큰지를 확인하는 조건식입니다.
만약 price가 2만원 이상이라면, 즉 조건식이 참이라면 shipping_cost 변수를 0으로 설정합니다. 이 경우 "구입금액이 2만원 이상입니다."라는 메시지를 출력합니다.
반대로, price가 2만원보다 작거나 같다면, 즉 조건식이 거짓이라면 shipping_cost 변수를 3000으로 설정합니다. 이 경우 "구입금액이 2만원 이하입니다."라는 메시지를 출력합니다.
마지막으로, f-string을 사용하여 "회원님의 배송비는 {shipping_cost}원입니다."라는 메시지를 출력합니다.
실행 예시:

상품의 가격을 입력하시오.
구입금액 입력: 25000
구입금액이 2만원 이상입니다.
회원님의 배송비는 0원입니다.
위의 예시에서는 구입금액이 2만원 이상이므로 배송비는 0원입니다.

상품의 가격을 입력하시오.
구입금액 입력: 15000
구입금액이 2만원 이하입니다.
회원님의 배송비는 3000원입니다.
위의 예시에서는 구입금액이 2만원 이하이므로 배송비는 3000원입니다.

3. 근무시간에 따라 초과근무와 정상근무를 판별하는 코드

- 코드 work_hour > 72

4. 짝수와 홀수를 판별하는 코드
- 사용변수 : _num_
- 수식 : (x % 2) !=0:

```python
print("정수를 입력하시오.")  # "정수를 입력하시오." 메시지를 화면에 출력합니다.
_num_ = int(input("정수 입력"))  # 사용자로부터 정수를 입력받아 정수형으로 변환하여 _num_ 변수에 저장합니다.

if (_num_ % 2) != 0:  # 입력된 정수가 홀수인지 확인합니다.
    print(f'입력하신 {_num_}은 홀수입니다.')  # 입력된 정수가 홀수일 경우 "입력하신 [입력된 정수]은 홀수입니다." 메시지를 출력합니다.
else:
    print(f'입력하신 {_num_}은 짝수입니다.')  # 입력된 정수가 짝수일 경우 "입력하신 [입력된 정수]은 짝수입니다." 메시지를 출력합니다.
```

위 코드는 사용자로부터 정수를 입력받아, 입력된 정수가 홀수인지 짝수인지를 판별하여 결과를 출력하는 코드입니다.

print("정수를 입력하시오.")는 사용자에게 "정수를 입력하시오."라는 메시지를 출력합니다.
_num_ = int(input("정수 입력"))는 사용자로부터 정수를 입력받아 정수형으로 변환하여 _num_ 변수에 저장합니다.
if (_num_ % 2) != 0:는 입력된 정수가 홀수인지를 확인하는 조건문입니다. 입력된 정수를 2로 나눈 나머지가 0이 아닌 경우, 즉 홀수인 경우 조건이 참이 됩니다.
print(f'입력하신 {_num_}은 홀수입니다.')는 입력된 정수가 홀수일 경우 "입력하신 [입력된 정수]은 홀수입니다."라는 메시지를 출력합니다.
else:는 조건식이 거짓인 경우를 다루는 코드 블록을 정의합니다.
print(f'입력하신 {_num_}은 짝수입니다.')는 입력된 정수가 짝수일 경우 "입력하신 [입력된 정수]은 짝수입니다."라는 메시지를 출력합니다.
위 코드를 실행하면 사용자로부터 정수를 입력받고, 입력된 정수가 홀수인지 짝수인지를 판별하여 결과를 출력합니다.

5. 사용자의 이름과 나이를 물어보고, 답변하는 코드
- 사용하는 변수: name, age
- 출력할 내용 :
  - 성명
  - 나이
  - (1) 20이하 : 프로그래밍을 배우는 최적의 나이입니다.
  - (2) 20이상 : 아직도 늦지 않았습니다.

```python
print("이름과 나이를 입력하시오.")  # "이름과 나이를 입력하시오." 메시지를 화면에 출력합니다.
name = input("이름: ")  # 사용자로부터 이름을 입력받아 name 변수에 저장합니다.
age = int(input("나이: "))  # 사용자로부터 나이를 입력받아 정수형으로 변환하여 age 변수에 저장합니다.

if age >= 20:  # 입력된 나이가 20 이상인지 확인합니다.
    print(f' 프로그래밍을 배우는 최적의 나이입니다. ')  # 나이가 20 이상일 경우 "프로그래밍을 배우는 최적의 나이입니다." 메시지를 출력합니다.
else:
    print(f'아직도 늦지 않았습니다.')  # 나이가 20 미만일 경우 "아직도 늦지 않았습니다." 메시지를 출력합니다.
```

위 코드는 사용자로부터 이름과 나이를 입력받고, 입력된 나이에 따라 프로그래밍을 배우는 최적의 나이인지를 판별하여 결과를 출력하는 코드입니다.
