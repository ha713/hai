```python
#5명의 성적을 입력을 받아 평균을 내는 프로그램을 만든다.
#반복문과 조건문을 사용한다.
#변수는  total, counter, grade, average
total =  0          #  변수 total은 0으로 선언 하였다.
counter = 1             # 변수 counter은 1로 선언 하였다.
while  counter <= 10:      # 반복문 whil을 사용하여 counter가 10보다 작거나 같을때 다음 내부문장으로 넘어간다.
            grade = int(input("성적을 입력하시오.:")) # 변수 grade에 int(정수형 변환 ) 와 사용자로부터 입력을 받는 input을 사용하여 정수를 받는다.
            total = grade + total           # 변수 total의  성적 grade와 변수 toal을 합한 값을  선언한다.
            counter = counter + 1        # 변수 counter는 초기 선언한 counter의 값에 실행할 때 마다 1씩 추가된다.
average = total/10          # 변수 averge에 반복문에서 계산한 total 값을 10으로 나눈다.
print(average)
# write by  김하영
```

total = 0: 변수 total을 0으로 초기화합니다. 이 변수는 입력받은 성적을 누적하여 저장할 때 사용됩니다.
counter = 1: 변수 counter을 1로 초기화합니다. 이 변수는 현재까지 입력받은 성적의 개수를 세는 역할을 합니다.
while counter <= 10:: while 반복문을 사용하여 counter가 10보다 작거나 같을 때까지 다음 내부 문장을 실행합니다. 이는 총 5명의 성적을 입력받기 위한 반복입니다.
grade = int(input("성적을 입력하시오.:")): input 함수를 사용하여 사용자로부터 성적을 입력받습니다. 입력받은 값은 int 함수를 사용하여 정수형으로 변환한 뒤, grade 변수에 저장됩니다.
total = grade + total: 입력받은 성적을 변수 total에 누적하여 저장합니다.
counter = counter + 1: counter 변수에 1을 더하여 다음 입력을 받을 준비를 합니다.
average = total/10: 입력받은 성적의 총합인 total을 10으로 나누어 평균을 계산하여 average 변수에 저장합니다.
print(average): 계산된 평균을 출력합니다.
위 코드를 실행하면 사용자로부터 5명의 성적을 입력받고, 입력받은 성적의 평균이 출력됩니다.
